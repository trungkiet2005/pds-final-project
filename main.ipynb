## 1.1. Chủ đề của dữ liệu

Dữ liệu này về **sức khỏe tinh thần tại nơi làm việc**, đặc biệt trong ngành công nghệ. Dataset chứa thông tin từ các cuộc khảo sát về trải nghiệm sức khỏe tinh thần của nhân viên, bao gồm:
- Lịch sử sức khỏe tinh thần cá nhân và gia đình
- Thái độ và hành vi tìm kiếm điều trị
- Các yếu tố liên quan đến công việc (căng thẳng, hứng thú công việc, v.v.)
- Triệu chứng và dấu hiệu (thay đổi tâm trạng, căng thẳng, khó khăn đối phó)

Đây là một vấn đề thực tế quan trọng trong xã hội hiện đại, đặc biệt trong ngành công nghệ nơi áp lực công việc cao.

```python
# 1. TIỀN XỬ LÝ: Chuyển Timestamp sang các cột số (Feature Engineering)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# Tạo cột 'Survey_Hour' (Giờ trong ngày: 0-23) - Đây là biến Numerical/Ordinal
df['Survey_Hour'] = df['Timestamp'].dt.hour

# Tạo cột 'Year' (Năm khảo sát) - Đây là biến Numerical/Discrete
df['Survey_Year'] = df['Timestamp'].dt.year

# Tạo cột 'Days_Elapsed' (Số ngày từ khi bắt đầu khảo sát) - Đây là biến Continuous Numerical
start_date = df['Timestamp'].min()
df['Days_Elapsed'] = (df['Timestamp'] - start_date).dt.days

print("PHÂN TÍCH CÁC BIẾN SỐ TỪ TIMESTAMP")

print("\nA. PHÂN TÍCH SURVEY_HOUR (GIỜ TRONG NGÀY)")

num_col = 'Survey_Hour'
stats = df[num_col].describe()
print(f"\n[1] Distribution & Central Tendency")
print(f"Mean (Trung bình): {stats['mean']:.2f}")
print(f"Median (Trung vị): {stats['50%']:.2f}")
print(f"Std Dev (Độ lệch chuẩn): {stats['std']:.2f}")

print(f"\n[2] Range & Outliers")
print(f"Min: {stats['min']}")
print(f"Max: {stats['max']}")

Q1 = df[num_col].quantile(0.25)
Q3 = df[num_col].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df[num_col] < lower_bound) | (df[num_col] > upper_bound)]
print(f"Số lượng Outliers (IQR method): {len(outliers)}")
print(f"Ngưỡng: [{lower_bound:.2f}, {upper_bound:.2f}]")

# Vẽ biểu đồ Survey_Hour
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(df[num_col].dropna(), kde=True, bins=24, color='skyblue')
plt.title('Distribution of Survey Hour', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day (0-23)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=df[num_col], color='lightgreen')
plt.title('Box Plot of Survey Hour', fontsize=12, fontweight='bold')
plt.xlabel('Hour of Day')

plt.tight_layout()
plt.show()


print("\nB. PHÂN TÍCH SURVEY_YEAR (NĂM KHẢO SÁT)")

year_col = 'Survey_Year'
year_stats = df[year_col].describe()
print(f"\n[1] Distribution & Central Tendency")
print(f"Mean (Trung bình): {year_stats['mean']:.2f}")
print(f"Median (Trung vị): {year_stats['50%']:.2f}")
print(f"Mode (Giá trị phổ biến nhất): {df[year_col].mode()[0]}")
print(f"Std Dev (Độ lệch chuẩn): {year_stats['std']:.2f}")

print(f"\n[2] Range")
print(f"Min Year: {int(year_stats['min'])}")
print(f"Max Year: {int(year_stats['max'])}")
print(f"Span: {int(year_stats['max']) - int(year_stats['min'])} years")

print(f"\n[3] Distribution by Year")
year_counts = df[year_col].value_counts().sort_index()
for year, count in year_counts.items():
    print(f"Year {int(year)}: {count:,} responses ({count/len(df)*100:.2f}%)")

# Vẽ biểu đồ Survey_Year
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
year_counts.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Distribution of Survey Year', fontsize=12, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Responses')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
sns.boxplot(x=df[year_col], color='lightcoral')
plt.title('Box Plot of Survey Year', fontsize=12, fontweight='bold')
plt.xlabel('Year')

plt.tight_layout()
plt.show()
```
